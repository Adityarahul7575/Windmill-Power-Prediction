```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score as score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df_train=pd.read_csv('/content/drive/MyDrive/Hackathon/HackerEarth/train.csv')
df_train
```

```python
df_test=pd.read_csv('/content/drive/MyDrive/Hackathon/HackerEarth/test.csv')
df_test
```

```python
df_train.columns
```

```python
df_train.info()
```

```python
df_train.turbine_status.value_counts()
```

```python
categorical_features = ['turbine_status','cloud_level']
```

```python
df_train.dropna(  axis=0, how='any', thresh=None, subset=None, inplace=True)
```

```python
df_train["turbine_status"]=pd.factorize(df_train.turbine_status)[0]
df_train["cloud_level"]=pd.factorize(df_train.cloud_level)[0]
```

```python
df_train.info()
```

```python
df_train=df_train.drop(labels=['tracking_id','datetime'],axis=1)
```

```python
df_train.dropna(  axis=0, how='all', thresh=None, subset=None, inplace=True)
```

```python
train, test = train_test_split(df_train, test_size=0.2,shuffle=True,random_state=56)
test.head(5)
```

```python
a=train.drop(labels='windmill_generated_power(kW/h)',axis=1)
xtrain=a 
ytrain=train['windmill_generated_power(kW/h)']
b =test.drop(labels='windmill_generated_power(kW/h)',axis=1)
xtest=b 
ytest=test['windmill_generated_power(kW/h)']
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(xtrain,ytrain )

train_score=score(ytrain,xgb.predict(xtrain))

test_score=score(ytest,xgb.predict(xtest))

print(train_score)

print(test_score)
```

```python
CBC= CatBoostRegressor(cat_features=categorical_features)
CBC.fit(xtrain,ytrain )

train_score=score(ytrain,CBC.predict(xtrain))

test_score=score(ytest,CBC.predict(xtest))

print(train_score)

print(test_score)
```
